```python
import pandas as pd
```

```python
df = pd.read_csv('data/interactive_data.csv')
df.head(5)
```

```python
#pre-processing: This removes rows where at least one column contains "None selected".
filtered_df = df[~df.eq('None selected').any(1)]
filtered_df.head(10)
```

```python
deaths = filtered_df['Deaths'].sum()
suicides = filtered_df[filtered_df['Intent'] == 'Suicide']['Deaths'].sum()
suicide_percentage = (suicides / deaths) * 100
print(f"The percentage of deaths due to suicide is {suicide_percentage:.2f}%")
```

```python
male_suicides = filtered_df[
                    (filtered_df['Intent'] == 'Suicide') & 
                    (filtered_df['Gender'] == 'Male')
                ]['Deaths'].sum()
male_suicide_percentage = male_suicides / suicides * 100
print(f"The percentage of male suicide is {male_suicide_percentage:.2f}%")
```

```python
homicides = filtered_df[filtered_df['Intent'] == 'Homicide']['Deaths'].sum()
homicides_percentage = homicides / deaths * 100
print(f"The percentage of deaths due to homicide is {homicides_percentage:.2f}%")
```

```python
young_black_male_homicides = filtered_df[
                            (filtered_df['Intent'] == 'Homicide') & 
                            (filtered_df['Gender'] == 'Male') &
                            (filtered_df['Age'] == '15 - 34') &
                            (filtered_df['Race'] == 'Black')
                        ]['Deaths'].sum()
young_male_homicides = filtered_df[
                            (filtered_df['Intent'] == 'Homicide') & 
                            (filtered_df['Gender'] == 'Male') &
                            (filtered_df['Age'] == '15 - 34')
                        ]['Deaths'].sum()
percentage = young_black_male_homicides / young_male_homicides * 100
print(f"The percentage of homicide victims who are black males in the age-group of 15--34 is {percentage:.2f}%")
```

```python
female_homicides = filtered_df[
                            (filtered_df['Intent'] == 'Homicide') & 
                            (filtered_df['Gender'] == 'Female')
                    ]['Deaths'].sum()
female_homicides_percentage = female_homicides / homicides * 100
print(f"The percentage of femais {percentage:.2f}%")
```
